```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=sns.load_dataset('iris')
df.head()
sns.relplot(x='sepal_length' , y ='petal_length',data = df)
```

```python
sns.countplot(x='sepal_length',data=df)
```

```python
sns.kdeplot(x = 'sepal_length' , data = df , color = 'black')
```

```python
sns.displot(x = 'petal_length',kde=True,bins = 5 , data =df)
```

```python
sns.displot(x ='petal_length',kde=True,bins = 5 ,
hue = df['species'] , palette = 'Set3', data=df)
```

```python
sns.jointplot(x='sepal_length' , y ='petal_length',
data = df , kind = 'reg')
```

```python
sns.pairplot(df)
```

```python
corr = df.corr()
sns.heatmap(corr)
```

```python
sns.heatmap(corr, cmap=['red','green','blue'],
vmin = -.5 , vmax = 0.6,annot = True)
```
